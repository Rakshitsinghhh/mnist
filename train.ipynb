```python
import numpy as np
from neural_network import NeuralNetwork
from data_loader import load_mnist
X_train, y_train = load_mnist()

```

```python
nn = NeuralNetwork()
lr = 0.1

for epoch in range(5):
    correct = 0
    for i in range(1000):
        X = X_train[i:i+1]
        y = y_train[i]

        out = nn.forward(X)
        pred = np.argmax(out)

        if pred == y:
            correct += 1

        nn.backward(X, y, lr)

    print(f"Epoch {epoch+1} Accuracy: {correct/10}%")
```
